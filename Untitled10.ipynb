##Prepairing the data

```python
df.mainroad.value_counts()
```

```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Convert 'yes'/'no' to 1/0, fill NaN with 0, then convert to int
df[varlist] = df[varlist].replace({'yes': 1, 'no': 0}).astype(int)
```

```python
df[varlist].head()
```

##Dummy variable

```python
#Creating dummy variable for furnishing status
status = pd.get_dummies(df['furnishingstatus'])
status.head()
```

```python
status = status.applymap(lambda x: 1 if x is True else (0 if x is False else x))
```

```python
status.head()
```

```python
#concat the dummy with original data
df = pd.concat([df,status], axis = 1)
df.head()
```

```python
df.drop(['furnishingstatus'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.drop(['furnished'], axis = 1, inplace = True)
```

```python
df.head()
```

##Splitting into train and test

```python
df_train,df_test = train_test_split(df,train_size = 0.7,random_state = 100)
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.info()
```

##Rescalling  the Features
###min-max scalling

```python
#Normalization : (x-xmin)/(xmax-xmin)
scaler = MinMaxScaler()
num_vars = ['area','bedrooms','bathrooms','stories','parking','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
df_train['mainroad'].value_counts()
```

```python
#Heatmap
plt.figure(figsize = (16,10))
sns.heatmap(df_train.corr(),annot = True,cmap= 'YlGnBu')
plt.show()
```
